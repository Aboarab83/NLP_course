```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.text,df.spam,test_size=0.25)
```

```python
print(X_train[:5])
```

```python
# create your bage of words counter vectorizer

counter=CountVectorizer()
counter.fit(X_train.values)
X_train_count=counter.transform(X_train.values)
X_test_count=counter.transform(X_test.values)

### YOU DO NOT NEED TO CONVERT THEM TO ARRAY ,FOR THE MODEL PURPOSE , BUT FOR EASIER PRINTTIN THEM WE WILL CONVERT TO ARRAY
X_train_arr=X_train_count.toarray()
X_test_arr=X_test_count.toarray()

```

**Note:** for the countvectorizer the output will be in the from of:
coords :(email index,word index in vocabluary list )     value: no of frequency how much it appears in this particular email

```python
# we will use multnominal because it is text with word counts (bag of words)
from sklearn.naive_bayes import MultinomialNB
multi_model=MultinomialNB()
multi_model.fit(X_train_count,y_train)
print(multi_model.score(X_test_arr,y_test))
y_predict=multi_model.predict(X_test_arr)
for i in range(15,30):
  print(y_predict[i],y_test.iloc[i])
```
